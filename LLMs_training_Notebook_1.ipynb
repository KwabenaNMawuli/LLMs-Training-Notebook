```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer('all-miniLM-L6-v2')
chunks = ['Creating embeddings using the sentence transformers module is really fun for Jack and Jill',
          'I have no knowledge at all about the use of RAG architectures',
          'Jack and Jill went up the hill']
embeddings = embedding_model.encode(chunks)
```

```python
similarities = embedding_model.similarity(embeddings, embeddings)
print(similarities)
print(embeddings)
```

```python
new_chunks = embedding_model.encode(["Jack can't do anything without Jill.", "There are no differences between the two."])

similarities = embedding_model.similarity(new_chunks, embeddings)
print(similarities)
```
